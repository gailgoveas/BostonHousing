```python
#importing modules

import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#Loading the dataframe

df = pd.read_csv('BostonHousing.csv')
df.head()
```

```python
df.dtypes
```

```python
df.columns[df.isna().any()].tolist()
```

#### Handpicking Features

```python
correlation_matrix = df.corr()
```

```python
fig = plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", annot_kws={"size": 6})
plt.savefig('heatmap.png', dpi=300)
```

```python
low_correlation_features = correlation_matrix[abs(correlation_matrix['medv']) < 0.5].index.tolist()
print(low_correlation_features)
```

```python
#Dropping columns to reduce multicollinearity
df.drop(["crim", "zn", "indus", "chas", "nox", "age", "dis", "rad", "tax", "b"], axis = 1, inplace=True)
```

```python
df.head()
```

```python
sns.pairplot(df)
plt.savefig('pairplot.png')
```

```python
#Dropping ptratio as it does not have a strong correlation with target
df.drop(["ptratio"], axis = 1, inplace=True)
```

```python
#Splitting data into training and testing set

X = df.drop("medv", axis = 1)
y = df["medv"]

housing_training_data, housing_test_data, housing_training_target, housing_test_target = train_test_split(
    X, y, test_size=0.40,random_state=42)
```

```python
# creating linear regression object
clf = linear_model.LinearRegression()
  
# training the model using the training sets
clf.fit(housing_training_data, housing_training_target)

#Performing 10 fold cross validation. 
neg_mse_scores = cross_val_score(clf, housing_training_data, housing_training_target, cv=10, scoring='neg_mean_squared_error', n_jobs = -1)

#Calculating RMSE scores
rmse_scores = np.sqrt(np.abs(neg_mse_scores))
```

```python
#calculating mean and standard deviation of RMSE values:

mean_rmse = np.mean(rmse_scores)
std_rmse = np.std(rmse_scores)

print('The mean RMSE is', round(mean_rmse,4) ,' and the standard deviation is', round(std_rmse,4))
```

#### Dimensionality Reduction using PCA

```python
#Loading the dataset into new dataframe
pca_df = pd.read_csv('BostonHousing.csv')
pca_df.head()
```

```python
#Applying Standard Scaler to dataset

scaler = StandardScaler()
X_features = pca_df.drop("medv", axis = 1)
y_label = pca_df["medv"]
scaled_features = scaler.fit_transform(X_features)
```

```python
# Creating principal components for features
pca = PCA()

# Fitting the PCA model to the standardized features
pca.fit(scaled_features)

#Determining the number of PCAS created
n_components = pca.n_components_
print("Number of principal components created", n_components)
```

```python
#Determining K value for PCA from looking at the sorted array of principal components.

explained_variance_ratio = pca.explained_variance_ratio_

cumulative_explained_variance_ratio = np.cumsum(explained_variance_ratio)

# Plotting the cumulative explained variance ratio
plt.plot(cumulative_explained_variance_ratio)
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.axhline(y=0.95, color='r', linestyle='--')
plt.show()
plt.savefig('pcaanalysis.png')

#Identifying the optimal KValue
optimal_kvalue = np.argmax(cumulative_explained_variance_ratio >= 0.95) + 1
print("The Optimal number of components based on 95% variance is", optimal_kvalue)
```

```python
# Creating principal components using optimal number of PCAS
pca = PCA(n_components = optimal_kvalue)

# Fitting the PCA model to the standardized features
pca_features = pca.fit_transform(scaled_features)
```

```python
#Splitting the data into training and testing
pca_training_data, pca_test_data, pca_training_target, pca_test_target = train_test_split(
    pca_features, y_label, test_size=0.40,random_state=42)

clf_pca = linear_model.LinearRegression()

#Performing 10 fold cross validation. 
pca_mse_scores = cross_val_score(clf_pca, pca_training_data, pca_training_target, cv=10, scoring='neg_mean_squared_error', n_jobs = -1)

#Calculating RMSE scores
pca_rmse_scores = np.sqrt(np.abs(pca_mse_scores))
```

```python
#calculating mean and standard deviation of RMSE values:

pca_mean_rmse = np.mean(pca_rmse_scores)
pca_std_rmse = np.std(pca_rmse_scores)

print('The mean RMSE is', round(pca_mean_rmse,4) ,' and the standard deviation is', round(pca_std_rmse,4))
```

```python
#Visualising the RMSE values of both models

fig, ax = plt.subplots()
sns.boxplot(data=[rmse_scores, pca_rmse_scores], palette='Pastel1', width=0.3, ax=ax)
ax.set_xticklabels(['Handpicked Features', 'PCA Method'])
ax.set_ylabel('RMSE')
plt.show()
plt.savefig('boxplot.png')
```

```python
ax = sns.barplot(x=['Handpicked Features', 'PCA Feature'], y=[mean_rmse, pca_mean_rmse], palette='Pastel1')
ax.set_xlabel('Models')
ax.set_ylabel('RMSE')
ax.set_title('Comparison of RMSE of two Models')
plt.show()
plt.savefig('barplot.png')
```
